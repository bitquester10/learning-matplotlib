# Learning matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Simple Example
Matplotlib graphs the data on `Figure`, each of which contains one or more `Axes`. An `Axes` is an area where points can be specified in terms of `x` and `y` coordinates. The simplest way of creating a `Figure` is using `pyplot.subplots`.

```python
# Create a figure containing a single axes
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])  # Plot some data on the axes.
```

## Using pylot instead of Axes
See [Matplotlib Application Interfaces (APIs)](https://matplotlib.org/stable/users/explain/figure/api_interfaces.html#api-interfaces) for an explanation of the trade-off of the supported user APIs.

See the tutorial: https://matplotlib.org/stable/tutorials/pyplot.html#sphx-glr-tutorials-pyplot-py

```python
# plot numbers on x-y plane. 
# If a single list or array is provided to plot, it will assume it to be y values and automatically generate the x values.
# The x values will start from 0
plt.plot([1, 2, 3, 4])
plt.ylabel('some numbers')
plt.show()
```

```python
# plotting with both x and y data
# the last argument is the color and format. 'ro' is red circles.
# plt.axis defines the range of the axes (x-min, x-max, y-min, y-max)
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
```

```python
# multiple plots and passing numpy arrays
t = np.arange(16)
# red dashes, blue squares, green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
# Plotting with keyword strings.
# With structured arrays, like dictionaries, you can pass the keys

# Scatter plot
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['s'] = np.abs(data['a'] - data['b'])

# c is color, s is size
plt.scatter('a', 'b', c='c', s='s', data=data)
# Label the x and y axis
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

## Plot using categorical variables

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]
plt.figure(figsize=(9, 3)) # width, height = 9, 3 inches
# 131: the three digits are nrows, ncols, index.
# This can also be written as (1, 3, 1 )
plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

## Controlling line properties

```python
x = np.arange(5)
y = x**2
plt.plot(x, y, linewidth=2.0)
plt.plot(x, x**3, linewidth=4.0)
plt.show()
```

```python
line1, line2= plt.plot(x, y, '-', x, x**3, '-')
# Turn on antialiasing. It is a technique to smooth the graph/edges.
line1.set_antialiased(False) 
plt.show()
```

```python
# Set multiple properties using setp
lines = plt.plot(x, y, x, x**3)
plt.setp(lines, color='r', linewidth=2.0)
plt.show()
```
